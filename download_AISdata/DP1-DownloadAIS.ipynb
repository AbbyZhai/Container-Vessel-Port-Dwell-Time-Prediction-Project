```python
import requests
import wget 
from datetime import timedelta, date
```

```python
def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)
```

```python
#start_dt = date(2022, 6, 1)
#end_dt = date(2022, 6, 30)
start_dt = date(2019, 7, 9)
end_dt = date(2019, 12, 31)
date_range = [dt.strftime("%Y_%m_%d") for dt in daterange(start_dt, end_dt)]
```

```python
# Change the year for your downlaod
base_url = "https://coast.noaa.gov/htdata/CMSP/AISDataHandler/2022/"
#base_url = "https://coast.noaa.gov/htdata/CMSP/AISDataHandler/2021/"
#base_url = "https://coast.noaa.gov/htdata/CMSP/AISDataHandler/2020/"
#base_url = "https://coast.noaa.gov/htdata/CMSP/AISDataHandler/2019/"
```

```python
files_url = [base_url +'AIS_' + dt + '.zip' for dt in date_range]
```

```python
download_dir = 'data/raw_ais_2022/'
#download_dir = 'assets/raw_ais/'
for f in files_url:
    #response = requests.get(f)
    file_name = f.split('/')[-1]
    #open('assets/'+file_name, 'wb').write(response.content)
    response = wget.download(f, download_dir +file_name)
```
